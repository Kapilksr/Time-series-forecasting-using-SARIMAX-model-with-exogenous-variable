## Import the model

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
len(df1)
```

#### train test split

```python
train= df1.iloc[:436]
test = df1.iloc[436:]
```

```python
model = SARIMAX(train['total'],exog=train[['holiday']],order = (0,0,1),seasonal_order=(2,0,0,7),enforce_invertibility=False)
```

#### fit the model

```python
results=model.fit()
```

```python
results.summary()
```

```python
start = len(train)
end = len(train)+len(test)-1
```

#### predict values for testing

```python
predictions=results.predict(start,end,exog=test[['holiday']]).rename('SARIMAX Pred')
```

```python
test['total']
```

```python
predictions
```

## plot the test values with the predicted values

```python
ax=test['total'].plot(figsize=(16,8),legend=True)
predictions.plot(legend=True)
ax.autoscale(axis='x',tight=True)

for day in test[test['holiday']==1].index:
    ax.axvline(x=day,color='black',alpha=0.8);
```

#### also check for rmse

```python
from statsmodels.tools.eval_measures import rmse
```

```python
rmse(test['total'],predictions)
```

```python
test['total'].mean()
```

```python
predictions.mean()
```
